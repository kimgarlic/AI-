### 파일열기 : open 함수

```python
f = open('데이터전처리1/file.txt', 'rt', encoding = 'UTF8')
# 파일명 -> open모드(r:읽기 + t:텍스트) -> 인코딩(UTF8 -> 한글취급)

for line in f : # 파일의 한 줄씩 처리하기
    print(line, end="")
```

<메모장><br>

r : 읽기모드<br>
w : 쓰기모드 (기존 내용은 사라짐, close해야 변경 내용이 반영a : 추가모드 (내용 추가, close해야 변경 내용이 반영)<br>
t : 텍스트모드 (텍스트를 기록할 때 사용)<br>
b : 바이너리모드 (바이트 단위로 기록할 때)<br>
x : 쓰기모드 (파일 생성 후 쓰기 모드로 열음)<br><br>

open 모드에 +를 포함하여 기존 파일 내용을 읽고 쓰는 방식으로 변경 가능<br>
rt+ : 텍스트 읽기 쓰기 모드 (기존 파일의 맨 뒤에서부터 내용을 씀)<br>
wt+ : 텍스트 읽기 쓰기 모드 (기존 파일의 내용을 지우고 처음부터 씀)<br>

###자동으로 파일 열고 닫기: with 문

```python
with open("데이터전처리1/proverb.txt", 'r', encoding = 'UTF8', newline = '') as filereader :
    for row in filereader :
        print(row.strip())

# open 함수에 with 문을 사용하면 파일 열기 닫기가 자동으로 처리 됨.
# strip() : 문자열의 양쪽 끝에서 공백(스페이스, 탭, 줄 바꿈 문자 등)을 제거
```

###여러 파일 읽기: os와 glob 모듈
- 폴더 내의 텍스트 파일 여러개를 모두 읽어서 출력

```python
import os
import glob

for file in glob.glob(os.path.join('*.txt')) :
    with open(file, 'r', newline = '', encoding = 'UTF8') as filereader :
        for row in filereader :
            print("{}".format(row.strip()))
```

### 문자열 공백/문자 제거: strip, lstrip, rstrip 함수

```python
b1 = " 작은 일에 성실한 사람은   큰일에도 성실하고 "
b1.strip()     # strip() : 앞뒤의 불필요한 공백을 제거
```

```python
text = "....작은 일에 성실한 사람은, 큰일에도 성실하고...."
print(text.lstrip('.'))     # 문자열 왼쪽의 '.' 제거
print(text.rstrip('.'))     # 문자열 오른쪽의 '.' 제거
print(text.strip('.'))      # 문자열 양쪽의 '.' 제거
```

<메모장><br>

strip() : 문자열 중간 공백은 제거되지 않음, 양쪽에 있는 공백만 제거<br>
lstrip() : 문자열 기준으로 왼쪽 공백이나 지정된 문자 제거<br>
rstrip() : 문자열 기준으로 오른쪽 공백이나 지정된 문자 제거<br>

###문자열 분리: split 함수

```python
text = "....작은 일에 성실한 사람은, 큰일에도 성실하고...."

# 결과는 리스트에 저장됨.
print(text.split(", "))

# split 요소별로 출력해보기
a = text.split(", ")
cnt = 1
for i in a :
    print(f'{cnt}번째 요소 : {i}')
    cnt += 1
```

- 코드가 간단하고 메모리가 절약됨.
- (lambda 매개변수 : 처리)(인수값)

```python
# 람다함수 미사용 시

def getSum(a, b) :
     x = a + b
     return x

print(getSum(10, 20))
```

```python
# 람다함수 사용 시

print((lambda x, y : x+y)(10, 20))
```

```python
# 리스트의 값을 정수로 변환하는 람다함수

# 람다함수 미사용 시

aa = [1.0, 34.7, 29.45, 22.4]

for i in range(len(aa)) :
    aa[i] = int(aa[i])

print(aa)
```

```python
# ★★★★★ (많이 사용)
# map 함수 : map()

aa = [1.0, 34.7, 29.45, 22.4]

list(map(int, aa))
# map(처리할 함수, 처리될 대상) -> list형으로 형변환
```

```python
# 람다함수 사용 시 (+ map 함수)

res = list(map(lambda x : int(x), aa))
res
```

```python
# 람다함수 사용 시 (+ map 함수)

res = list(map(lambda x : x*2, aa))
res
```

```python
# 성별을 숫자로 바꾸기
# sex 컬럼의 값 : 'male', 'female' --> 남자 1, 여자 0으로 변경하여 gender 컬럼으로 추가

titanic_df1 = titanic_df[['survived', 'pclass', 'name', 'sex', 'age', 'sibsp', 'parch', 'fare']]

### FOCUS (실습 부분) ###

# 방법 1

res = list(map(lambda x : 1 if x == 'male' else 0, titanic_df1['sex']))
titanic_df1['gender'] = res
# map() -> 적용될 처리 : lambda x : 1 if x == 'male' else 0
#                        (삼항연산자 사용 : 만약에 x값이 'male'이라면 결과값은 1, 아니라면 결과값은 0)
#          적용될 대상 : titanic_df1['sex']
#                        (df의 'sex' 컬럼값들을 순회하여 x에 대입)
# map() 결과를 -> list형화 한 후의 결과를 'gender'라는 새로운 컬럼값으로 대입

titanic_df1
# 'sex' 컬럼값이 'male'일 때는 'gender'컬럼이 1 // 'female'일 때는 0으로 'gender'컬럼값을 가지는 것을 확인 가능
```

### groupby 함수
- Group By는 데이터를 그룹화하고 그룹화된 데이터에 대한 연산을 수행하는 기능
- Group By를 사용하면 데이터를 더 쉽게 분석
- df.groupby('그룹화할 열 이름')

```python
# group by의 의해

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {'학년': ['1학년', '1학년', '2학년', '2학년', '3학년', '3학년', '3학년'],
        '성별': ['남', '여', '남', '여', '남', '여', '여'],
        '성적': [80, 90, 70, 60, 85, 95, 75]}

df = pd.DataFrame(data)
df

# 딕셔너리형 -> 데이터프레임을 생성
```

```python
# 학년별 성적 평균

grouped = df.groupby('학년')
# '학년' 컬럼값을 기준으로 그룹화한 df => grouped

result = grouped[['성적']].mean()
# 그룹화한 df -> '성적' 컬럼에 대한 평균값을 집계

result
```

```python
# 참고 사항

result = grouped['성적'].mean()
result
# 1차원 리스트로 컬럼명을 지정했기 때문에 시리즈로 출력된다.
# 출력의 통일성(데이터프레임으로 출력)을 위해 기준 컬럼이 1개여도 2차원 리스트로 지정한다.
```

```python
# 학년별 성적 통계

grouped = df.groupby('학년')
result = grouped['성적'].agg(['mean', 'min', 'max', 'count'])
# agg : 여러개의 집계함수를 적용하는 수단 (적용할 집계함수들을 리스트로 정리)
# 집계함수가 다수개 이므로 '성적' 컬럼 선택을 1차원 리스트로 지정해도
# 출력결과는 시리즈가 아닌 데이터프레임으로 됨.

result
```

```python
# 그룹화 기준 컬럼이 다수 개일 때는 어떻게?

# 학년별 성별 성적 통계

grouped = df.groupby(['학년', '성별'])
# 기준컬럼이 다수 개면 리스트형으로 제시

result = grouped['성적'].agg(['mean', 'min', 'max', 'count'])
result
# 학년 그룹화 내에 성별 그룹화도 이뤄졌음을 확인 가능.
```

### cut 함수
- 데이터를 구간 길에 따라 나눌때
- cut( arrary, bins, labels )
  - bins = 3 :  데이터의 최소 최대 구간으로 3등분
  - bins - [ 0, 30, 70, 100 ] :  왼쪽은 포함하지않고 오른쪽은 포함 , 첫구간은 시작점 포함  // 0~ 30 , 31 ~ 70, 71~ 100
  - labels : 구간의 이름

-----------
사용 : 연령대별, 성적구간별, 소득구간별, 사용시간대별, ...

```python
import random

a = [0, 10, 30, 22, 55, 31, 66, 70, 71, 89, 75, 90, 91, 100]

df = pd.DataFrame(a, columns=['num'])
df
```

```python
# 등급을 추가 (미완성본)

df['grade'] = pd.cut(df['num'], bins=[0, 30, 70, 100], labels=['C', 'B', 'A'])
# pd.cut() - df['num'] : cut이 될 대상
#          - bins : 계급표시 (★ 좌측값은 초과처리 // 우측값은 이하처리)
#                   if) 'num' 컬럼값이 0이면, 어떠한 계급에도 해당되지 않으니 'grade' 컬럼값은 결측값으로 지정
#          - labels : 계급에 해당할 경우 치환되는 라벨값
#                     (0 ~ 30 ->  C  // 31 ~ 70 -> B // 71 ~ 100 -> A)

df
```

```python
# 등급을 추가 (완성본)

df['grade']=pd.cut( df['num'], bins=[0,30,70,100], labels=['C','B','A'], \
                   include_lowest=True)
# 기존 bins 규칙 : 좌측값은 초과처리 // 우측값은 이하처리
#                  첫 계급은 0 초과 ~ 30 이하 => 만약 'num' 컬럼값이 0이라면 등급은 어떻게 결정?
# include_lowest = True : 첫 계급은 좌측값을 초과처리 -> 이상처리
#                         첫 계급은 0 이상 ~ 30 이하

df
```
